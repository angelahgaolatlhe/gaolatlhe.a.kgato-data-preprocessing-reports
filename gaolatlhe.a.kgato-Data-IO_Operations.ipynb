```python
world_gapminder = pd.read_csv('data/gapminder.tsv' , sep = "\t")

world_gapminder.head(10)
```

```python
# list all the countries in our dataframe to check if Angola is in the list

world_gapminder["country"].unique()
```

```python
# How many countries are there?

world_gapminder["country"].nunique()
```

```python
Angola = world_gapminder[world_gapminder["country"] == "Angola"]

Angola
```

```python
above_40 = Angola[Angola["lifeExp"] > 40]

above_40.head(10)
```

```python
year_against_population = above_40[["year","pop"]]

year_against_population
```

```python
import matplotlib.pyplot as plt

plt.plot(year_against_population["year"], year_against_population["pop"], label = "population")
plt.xlabel("year")
plt.ylabel("pop")
plt.legend()
plt.title("Life expectancy more than 40 in Angola")
```

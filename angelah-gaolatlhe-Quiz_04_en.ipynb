 ## Suppose you are given the following date to parse: `'2025 KEUR 12***04'` . Run the code below
 
  

```python
saly_date_coerce = pd.to_datetime('2025 KEUR 12***04', format='%Y KEUR %m--%d',
                   errors='coerce')
saly_date_coerce
```

## 6. We have passed in `ignore` as parameter. Describe what exactly happen?

```python

saly_date_ignore = pd.to_datetime('2025 KEUR 12***04', format='%Y KEUR %m--%d',
                   errors='ignore')
saly_date_ignore
```

The separator between month and day is "***" instead of "--%" listed in the format. When we use coerce, it passed NaT because the format was incorrect but when we use ignore, the function simply ignores the error and returns the original entry.

## 7. We have passed in `raise` as parameter. Describe what exactly happen?

```python

saly_date_raise = pd.to_datetime('2025 KEUR 12***04', format='%Y KEUR %m--%d',
                   errors='raise')
saly_date_raise
```

Raise causes an error message to be displayed because the format in the data entered is not the same as the format given inside the paranthesis.

### ✏️ **Practice Exercise 1: African Literature**

Create dataclass instances for these famous African books:
1. "Things Fall Apart" by Chinua Achebe (Nigeria), 209 pages, published 1958
2. "Half of a Yellow Sun" by Chimamanda Ngozi Adichie (Nigeria), 433 pages, published 2006
3. "Nervous Conditions" by Tsitsi Dangarembga (Zimbabwe), 204 pages, published 1988

**Tasks:**
- Create three Book instances with the data above
- Store them in a list called `african_books`
- Print all three books
- Calculate and print the average number of pages
- Find and print the oldest book

**Hint:** Use a for loop to iterate through the list!

```python
from dataclasses import dataclass

@dataclass
class Book:
    title: str
    author: str
    country: str
    pages: int 
    year_published : int
```

```python
book_1 = Book(title = "Things Fall Apart", author = "Chinua Achebe", country = "Nigeria", pages = 209, 
              year_published = 1958 )
book_1
```

```python
book_2 = Book(title = "Half of a Yellow Sun", author = "Chimamanda Ngozi Adichie", country = "Nigeria", pages = 433, 
              year_published = 2006 )
book_2
```

```python
book_3 = Book(title = "Nervous Conditions" , author = "Tsitsi Dangarembga", country = "Zimbabwe", pages = 204, 
              year_published = 1988 )
book_3
```

```python
african_books = [book_1, book_2, book_3]

african_books
```

```python
average = (book_1.pages + book_2.pages + book_3.pages)/3

average
```

```python
if book_1.year_published < book_2.year_published and book_1.year_published < book_3.year_published:
    print(f"{book_1.title} is the oldest book")
elif book_2.year_published < book_1.year_published and book_2.year_published < book_3.year_published:
    print(f"{book_2.title} is the oldest book")
else:
    print(f"{book_3.title} is the oldest book")
```

### **Part 3: Multiple Perspectives on Same Entity**

**Exercise 5: The Baobab Tree (Senegal)**

The ancient baobab tree in a Senegalese village is 300 years old and 25 meters tall. Local tradition considers it a sacred meeting place, while ecologists study its role in the ecosystem.

- Create a `SacredTree` dataclass with cultural attributes
- Create an `EcologicalSpecimen` dataclass with scientific attributes
- Make one instance for the baobab tree using each perspective

```python
from dataclasses import dataclass

@dataclass
class SacredTree:
    name: str
    age: str
    purpose: str
```

```python
cultural_attributes = SacredTree(name = "Baobab tree", age = "very old tree", purpose = "Sacred meeting place")
cultural_attributes
```

```python
@dataclass
class EcologicalSpecimen:
    name: str
    age: int
    height: float
```

```python
scientific_attributes = EcologicalSpecimen(name = "Baobab tree", age = 300, height = 25)
scientific_attributes
```

**Exercise 6: Nigerian Entrepreneur**

Chinedu Okoro runs a Lagos-based tech startup with 15 employees. Outside work, he's a father of three who enjoys playing football on weekends.

- Create a `BusinessOwner` dataclass with professional attributes
- Create a `FamilyPerson` dataclass with personal attributes  
- Make one instance for Chinedu using each perspective

### **Part 4: Multiple Perspectives on Different Entities**

**Exercise 7: Ghanaian Education & Agriculture**

In Kumasi, Professor Kwame Asante has taught mathematics for 20 years and published 15 research papers. Meanwhile, cocoa farmer Akua Mensah harvests 2 tons of cocoa beans annually from her 10-acre farm.

- Create an `Academic` dataclass for the professor
- Create an `AgriculturalProducer` dataclass for the farmer
- Make appropriate instances for both people

```python
from dataclasses import dataclass

@dataclass
class Academic:
    name: str
    profession: str
    specialisation: str
    experience: int
    number_of_published_papers: int
```

```python
professor = Academic(name = "Kwame Asante", profession = "Professor", specialisation = "mathematics", 
                     experience = 20, number_of_published_papers = 15)
professor
```

```python
@dataclass
class AgriculturalProducer:
    name: str
    profession: str
    specialisation: str
    size_of_farm: float
    total_produce: float
```

```python
farmer = AgriculturalProducer(name = "Akua Mensah", profession = "Farmer", specialisation = "Cocoa", 
                              size_of_farm = 10, total_produce = 2)
farmer
```

**Exercise 8: Ethiopian Tourism & Conservation**
The Simien Mountains in Ethiopia attract 50,000 tourists yearly for hiking, while conservationists work to protect the endangered Ethiopian wolf with only 500 remaining in the wild.

- Create a `TouristDestination` dataclass for the mountains
- Create an `EndangeredSpecies` dataclass for the wolf
- Make appropriate instances for both

```python
from dataclasses import dataclass

@dataclass
class TouristDestination:
    name: str
    location: str
    number_of_tourists: int
```

```python
mountains = TouristDestination(name = "Simien mountains", location = "Ethiopia", number_of_tourists = 50000)
mountains
```

```python
@dataclass
class EndangeredSpecies:
    name: str
    number_remaining: int
```

```python
wolf = EndangeredSpecies(name = "Ethiopian Wolf", number_remaining = 500)
wolf
```

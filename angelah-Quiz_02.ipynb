### **Question 2: Create Dataclasses<span style="color:red">Score: 4 Marks</span>**

 Create two subclasses: `Frog` and `Parrot`, each with their own unique attributes (e.g., `Frog` could have `length: float`, `weight: float`, and `habitat: str`). Instantiate one object for each subclass.


```python
@dataclass
class Frog:
    name: str
    type_of_animal: str
    colour: str
    length: float
    weight: float
    habitat: str
    
Animal_1 = Frog(name = "Goliath frog", type_of_animal = "Frog", 
                length = 32, weight = 3.3, 
                colour = "Greenish-brown", habitat = "rivers")
Animal_1
```

```python
@dataclass
class Parrot:
    name: str
    type_of_animal: str
    colour: str
    life_expectancy: float
    special_quality: str
    
Animal_2 = Parrot(name = "African grey parrot", type_of_animal = "Bird",
                 colour = "gray", life_expectancy = 50, special_quality = "Can copy human speech" )
Animal_2
```

### **Question 4: Use Literal Types  <span style="color:red">Score: 2 Marks</span>**
**Task:** Modify the `Parrot` dataclass so that the `color` attribute can only be one of the following: `"gray"`, `"red"`, or `"white"`. Use Python’s `Literal` type to enforce this.

### **Question 4: Convert to Pydantic BaseModel <span style="color:red">Score: 5 Marks</span>**
**Task:** Rewrite the `Animal`, `Frog`, and `Parrot` classes using Pydantic’s `BaseModel` instead of dataclasses. Ensure all attributes are properly typed and include validation for the `Parrot`'s `color` attribute.

```python
class Animal(BaseModel):
    name: str
    type_of_animal: str
    colour: str
```

```python
class Frog(BaseModel):
    name: str
    type_of_animal: str
    colour: str
    length: float
    weight: float
    habitat: str
    
Animal_1 = Frog(name = "Goliath frog", type_of_animal = "Frog", 
                length = 32, weight = 3.3, 
                colour = "Greenish-brown", habitat = "rivers")
Animal_1
```

```python
class Parrot(BaseModel):
    name: str
    type_of_animal: str
    colour: Literal[str] = None
    life_expectancy: float
    special_quality: str
    
Animal_2 = Parrot(name = "African grey parrot", type_of_animal = "Bird",
                 colour = "gray", life_expectancy = 50, special_quality = "Can copy human speech" )
Animal_2
```
